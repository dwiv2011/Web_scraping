Next block is for extracting all the URLS of different articles

```python
title=[]
date=[]
summary=[]

for m in range(1,100):
    driver.get("https://www.moneycontrol.com/news/tags/companies.html/page-"+str(m)+"/")        
    # counter for number of articles in the page
    counter=driver.find_elements_by_xpath('//*[@id="cagetory"]/li')
    count=len(counter)
    #extracting the URLS of each article
    for k in range(0,count):
        #store the href
        try:
            ti=driver.find_element_by_xpath('//*[@id="newslist-'+str(k)+'"]/h2/a').get_attribute("title")
            title.append(ti)
        
            try:
                d=driver.find_element_by_xpath('//*[@id="newslist-'+str(k)+'"]/div/span').text
                date.append(d)
                
            except:
                d=driver.find_element_by_xpath('//*[@id="newslist-'+str(k)+'"]/span').text
                date.append(d)
            su=driver.find_element_by_xpath('//*[@id="newslist-'+str(k)+'"]/p').text
            summary.append(su)
        
        except:
            pass
        
        
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
data=pd.DataFrame(zip(title,summary,date),columns=["Title","Article","Date"])
data.to_csv("Money_control_scrape.csv")
```

```python
dtext=data['Article']

import nltk
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))

from textblob import TextBlob

sentimentspol=[]
sentimentssen=[]
for i in dtext:
    senti=TextBlob(i).correct()
    score=senti.sentiment
    #print(score)
    word=str(score).split('=')
    polarity=word[1].split(',')[0]
    sensitivity=word[2].split(')')[0]
    sentimentspol.append(polarity)
    sentimentssen.append(sensitivity)
print(sentimentspol)
```

```python
#nltk.download('brown')
nou=[]
art=[]
for i in dtext:
    art.append(i)
    s=" "
    for np in TextBlob(i).noun_phrases:
        s=s+" "+np
    nou.append(s)
    
```

```python
data2=pd.DataFrame(zip(art,sentimentspol,sentimentssen,nou),columns=["Article","Sentiment Score"," Subjectivity","Noun"])
Final_data=pd.merge(data,data2,on="Article",how="inner" )
```
